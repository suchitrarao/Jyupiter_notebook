This Note book is explaining various graphical representation of data in the form of charts by using Matplotlib library in python.
Charts are:
    1. Line chart
    2. Bar chart
    3. Pie chart

# Line Chart

```python
import numpy as np
import matplotlib.pyplot as plt

figure, axes = plt.subplots(nrows=1, ncols=2)
figure.tight_layout(pad=2.0)
plt.subplots_adjust(top=0.8)

#plot 1:
x = np.linspace(-5.5, 5.5, num=100)
y = np.sin(x)
plt.subplot(1, 2, 1)
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.title("Sine Wave")
plt.plot(x, y)

# plot 2:
ground_speed = np.random.choice([3, 5, 7, 9])
x1 = np.linspace(-2 * np.pi, 2 * np.pi, 1000);
y1 = ground_speed * (np.tan(x1))
plt.subplot(1, 2, 2)
plt.xlabel("Flight Path angle")
plt.ylabel("Altitude")
plt.title("Altitude change of aircraft")
plt.plot(x1, )
plt.ylim(-5, 5)


plt.suptitle("Scientific functions")
plt.show()
```

# Bar Chart

```python
import matplotlib.pyplot as plt
import numpy as np

Women = [115, 220, 420, 400]
Men = [120, 230, 510, 370]
  
n=4
r = np.arange(n)
width = 0.25

plt.bar(r, Women, color = 'm',
        width = width, edgecolor = 'black',
        label='Women')
plt.bar(r + width, Men, color = 'b',
        width = width, edgecolor = 'black',
        label='Men')
  
plt.xlabel("Year")
plt.ylabel("Number of students graduated")
plt.title("Graduation data")
  
# plt.grid(linestyle='--')
plt.xticks(r + width/2,['2018','2019','2020','2021'])
plt.legend()
  
plt.show()
```

# Pie Chart

```python
import matplotlib.pyplot as plt

labels = ['Samsung', 'Oppo', 'Apple', 'Tecno']
sizes = [25, 15, 35, 25]
explode = (0.1, 0.1, 0, 0)

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
ax1.axis('equal')

plt.legend(title="OS usage", loc="center right")
plt.show()
```
